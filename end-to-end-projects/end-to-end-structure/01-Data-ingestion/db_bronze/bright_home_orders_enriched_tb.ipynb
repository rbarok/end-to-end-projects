- bright home orders
- customer changes daily

```python
%sql
CREATE OR REPLACE TABLE end_to_end_retail.db_bronze.bright_home_orders_enriched
AS
SELECT 
    bho.order_id,
    bho.customer_id,
    bho.order_timestamp,
    bho.channel,
    bho.sku,
    bho.category,
    bho.qty,
    bho.unit_price,
    bho.discount_pct,
    bho.coupon_code,
    bho.total_amount,
    bho.region,
    bho.city,
    bho.order_date,
    -- bho.file_path,
    bho.ingestion_date,
    ccd.loyalty_tier,
    ccd.email,
    ccd.first_name,
    ccd.last_name,
    ccd.signup_date,
    ccd.source_subsidiary,
    ccd.operation,
    ccd.timestamp as customer_change_timestamp,
    nso.email as northstar_email,
    nso.first_name as northstar_first_name,
    nso.last_name as northstar_last_name,
    nso.loyalty_tier as northstar_loyalty_tier,
    nso.signup_date as northstar_signup_date
FROM end_to_end_retail.db_bronze.bright_home_orders bho
LEFT JOIN end_to_end_retail.db_bronze.customer_changes_daily ccd
    ON bho.customer_id = ccd.customer_id
LEFT JOIN end_to_end_retail.db_bronze.northstar_outfitters_orders nso
    ON bho.customer_id = nso.customer_id
QUALIFY ROW_NUMBER() OVER (
    PARTITION BY bho.order_id, bho.customer_id, bho.order_timestamp
    ORDER BY bho.ingestion_date DESC
 
) = 1
```
