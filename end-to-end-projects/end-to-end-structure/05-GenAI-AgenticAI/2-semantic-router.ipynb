```python
import os
from semantic_router import Route
from semantic_router.encoders import OpenAIEncoder
from semantic_router.layer import RouteLayer

# 1. DEFINE DESTINATIONS
# We define "Routes" which act as our destinations. 
# Each has a name and sample phrases (utterances) that describe its intent.
sales_route = Route(
    name="sales",
    utterances=[
        "how much does the premium plan cost?",
        "I want to buy a subscription",
        "is there a discount for annual billing?",
        "who do I talk to about a corporate license?"
    ],
)

support_route = Route(
    name="support",
    utterances=[
        "how do I reset my password?",
        "my account is locked",
        "I'm seeing an error 500 when I log in",
        "how do I update my email address?"
    ],
)

# 2. EMBED DESTINATIONS
# We choose an encoder (OpenAI, Cohere, HuggingFace, etc.) 
# to transform the utterances above into vector embeddings.
os.environ["OPENAI_API_KEY"] = "your-api-key-here"
encoder = OpenAIEncoder()

# 3. SEMANTIC SEARCH & ROUTING LAYER
# The RouteLayer stores the embeddings and performs the similarity search.
routes = [sales_route, support_route]
rl = RouteLayer(encoder=encoder, routes=routes)

def process_query(user_query: str):
    print(f"\nUser Query: {user_query}")
    
    # 4. PROCESS USER INPUT & 5. SEMANTIC SEARCH
    # The library embeds the input and calculates cosine similarity automatically.
    route_choice = rl(user_query)
    
    # 6. ROUTE THE QUERY
    if route_choice.name == "sales":
        return "Routing to: [Sales Knowledge Base Chain]"
    elif route_choice.name == "support":
        return "Routing to: [Customer Support Flow Chain]"
    else:
        return "Routing to: [General LLM / Default Fallback]"

# --- Testing the Router ---
print(process_query("I need to change my billing password"))
print(process_query("what are your pricing tiers?"))
print(process_query("how is the weather today?"))
```
