#### Landing layers, ideally would be S3 bucket; soon

- Meditations

```python
import pandas as pd

# 1. Read the entire file content into a string
with open('/Volumes/llm/rag/txt_vol/meditations.txt', 'r') as file:
    content = file.read()

# 2. Split the content by double newlines to separate blocks
blocks = content.strip().split('\n\n')

# 3. Create the DataFrame
df = pd.DataFrame(blocks, columns=['Content'])
df['id'] = ['meditations_' + str(i) for i in range(len(df))]

# 4. Write DataFrame to JSON file in the specified volume
df.to_json('/Volumes/end_to_end_genie_room/default/db_landing/meditations_blocks.json', orient='records', lines=True)
```
