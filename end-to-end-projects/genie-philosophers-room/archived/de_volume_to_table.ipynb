- Meditations

```python
# 1. Read the .txt files as binary
df = spark.read.format("binaryFile").load("/Volumes/llm/rag/txt_vol/meditations.txt")
```

```python
import pandas as pd

# 1. Read the entire file content into a string
with open('/Volumes/llm/rag/txt_vol/meditations.txt', 'r') as file:
    content = file.read()

# 2. Split the content by double newlines to separate blocks
# This creates a list where each element is a block of text
blocks = content.strip().split('\n\n')

# 3. Create the DataFrame
df = pd.DataFrame(blocks, columns=['Content'])
```

```python
spark_df = spark.createDataFrame(df)
spark_df.write.format("delta").mode("overwrite").saveAsTable("llm.rag.meditations_blocks")
```
