```python
%sql
CREATE OR REPLACE TABLE end_to_end_retail.db_silver.tbs_customer_ordered_products_geoloc (
  customer_id STRING PRIMARY KEY,
  ship_to_address STRING,
  clicked_items STRING,
  lon DOUBLE,
  lat DOUBLE,
  valid_from STRING,
  valid_to STRING,
  order_number STRING,
  ordered_products STRING
);

INSERT INTO end_to_end_retail.db_silver.tbs_customer_ordered_products_geoloc
SELECT 
  geo.customer_id,
  geo.ship_to_address,
  geo.clicked_items,
  geo.lon,
  geo.lat,
  geo.valid_from,
  geo.valid_to,
  prod.order_number,
  prod.ordered_products
FROM end_to_end_retail.db_silver.tbs_customer_address_clicks_geoloc geo
LEFT JOIN end_to_end_retail.db_landing.tbl_ordered_product prod
  ON geo.customer_id = prod.customer_id
WHERE prod.customer_id IS NOT NULL;
```
