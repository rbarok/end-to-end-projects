### Naming

```python
%sql
SET CATALOG end_to_end_retail;
USE SCHEMA db_landing;
SELECT current_catalog();
```

```python
catalog = spark.catalog.currentCatalog()
schema = spark.catalog.currentDatabase()
tables = spark.sql(f"SHOW TABLES IN {catalog}.{schema}").select("tableName")
table_names = [row.tableName for row in tables.collect()]
for table in table_names:
    new_table = f"tbl_{table}"
    spark.sql(f"ALTER TABLE {catalog}.{schema}.{table} RENAME TO {catalog}.{schema}.{new_table}")
```
